```python
import pandas as pd

# Muatkan data
df = pd.read_csv('Data/Cleaned/cleaned_data.csv')

# Tukarkan jawapan kepada nilai binari
df['Pelanggan_Setia'] = df['20. Will you continue buying at Starbucks?'].map({'Yes': 1, 'No': 0})
df
```

# UJIAN ANOVA (Melibatkan korelasi kualtitif ) Norminal dan ordinal

```python
#Analisis ANOVA untuk Kategori Umur vs Kesetiaan Pelanggan
#Kita akan menggunakan ANOVA satu hala untuk melihat jika terdapat perbezaan yang signifikan dalam kesetiaan pelanggan berdasarkan umur

from scipy.stats import f_oneway

# Buat subset data mengikut kategori umur dan bandingkan kesetiaan pelanggan
anova_age = df.groupby('2. Your Age')['Pelanggan_Setia']

# ANOVA untuk kesetiaan berdasarkan umur
f_val, p_val = f_oneway(*[anova_age.get_group(age) for age in anova_age.groups])
print(f'ANOVA Umur vs Kesetiaan Pelanggan: F-value = {f_val}, p-value = {p_val}')
```

```python
#Visualisasi untuk Umur vs Kesetiaan Pelang
#Gunakan boxplot untuk melihat perbezaan dalam kategori umur tanpa mapping ke nilai numerik.

sns.boxplot(x='Pelanggan_Setia', y='2. Your Age', data=df, order=['Below 20', 'From 20 to 29', 'From 30 to 39', 'From 40 to 49', '50 and above'])
plt.title('Umur vs Kesetiaan Pelanggan di Starbucks')
plt.xlabel('Kesetiaan Pelanggan')
plt.ylabel('Umur')
plt.show()
```

```python
#Analisis ANOVA untuk Pendapatan vs Kesetiaan Pelanggan
# Buat subset data mengikut kategori pendapatan
anova_income = df.groupby('4. What is your annual income?')['Pelanggan_Setia']

# ANOVA untuk kesetiaan berdasarkan pendapatan
f_val_income, p_val_income = f_oneway(*[anova_income.get_group(income) for income in anova_income.groups])
print(f'ANOVA Pendapatan vs Kesetiaan Pelanggan: F-value = {f_val_income}, p-value = {p_val_income}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Violin plot untuk Pendapatan vs Kesetiaan Pelanggan
sns.violinplot(
    x='Pelanggan_Setia', 
    y='4. What is your annual income?', 
    data=df, 
    order=['Less than RM25,000', 'RM25,000 - RM49,999', 'RM50,000 - RM74,999', 'RM75,000 - RM99,999', 'RM100,000 and above']
)

plt.title('Pendapatan vs Kesetiaan Pelanggan di Starbucks')
plt.xlabel('Kesetiaan Pelanggan')
plt.ylabel('Pendapatan')
plt.show()
```

# Analisis ciri-ciri dan kemudahan di Starbucks yang paling menarik perhatian pelanggan, seperti lokasi, suasana, kemudahan Wi-Fi, menu terhadap kesetian pelanggan

```python
from scipy import stats
# Membaca data
df = pd.read_csv('Data/Cleaned/cleaned_data.csv')

# Memilih kolom dengan iloc
wifi_quality = df.iloc[:, 15]  # Kolom WiFi quality
continue_buying = df.iloc[:, 19]  # Kolom Continue buying

# Membuat tabel kontingensi
#Karena satu kolom bersifat ordinal dan satu nominal, kita bisa menggunakan Chi-Squared test untuk analisis ini:
contingency_table = pd.crosstab(wifi_quality, continue_buying)

# Melakukan uji Chi-Squared
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("Chi-Squared Value:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)

```

Dengan p-value yang lebih besar dari 0.05, Anda tidak dapat menyimpulkan bahwa ada hubungan yang signifikan antara penilaian kualitas WiFi dan keputusan untuk melanjutkan pembelian di Starbucks. Mungkin ada faktor lain yang mempengaruhi keputusan pembelian yang tidak tercakup dalam analisis ini

```python
import pandas as pd
from scipy import stats

# Membaca data
df = pd.read_csv('Data/Cleaned/cleaned_data.csv')

# Memilih kolom dengan iloc
location = df.iloc[:, 7]  # Kolom lokasi
loyalty = df.iloc[:, 19]  # Kolom kesetiaan

# Membuat tabel kontingensi
contingency_table = pd.crosstab(location, loyalty)

# Melakukan uji Chi-Squared
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Menampilkan hasil
print("Chi-Squared Value:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
```

Dengan p-value 0.068, Anda dapat menyimpulkan bahwa ada indikasi hubungan antara lokasi dan kesetiaan pelanggan di Starbucks, meskipun tidak cukup kuat untuk dinyatakan signifikan pada level 0.05. Ini bisa menunjukkan bahwa lokasi memang dapat mempengaruhi keputusan pelanggan untuk tetap setia kepada merek.

```python
import pandas as pd
from scipy import stats  # Pastikan untuk mengimpor stats di sini

# Membaca data
df = pd.read_csv('Data/Cleaned/cleaned_data.csv')

# Memilih kolom dengan iloc
importance_sales_promotions = df.iloc[:, 13]  # Kolom pentingnya penjualan dan promosi
loyalty = df.iloc[:, 19]  # Kolom kesetiaan

# Membuat tabel kontingensi
contingency_table = pd.crosstab(importance_sales_promotions, loyalty)

# Melakukan uji Chi-Squared
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Menampilkan hasil
print("Chi-Squared Value:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
```

Hasil analisis ini menunjukkan bahwa ada indikasi hubungan antara seberapa penting penjualan dan promosi dalam keputusan pembelian dengan kesetiaan pelanggan di Starbucks, meskipun tidak cukup kuat untuk dianggap signifikan pada level 0.05. Anda mungkin ingin mempertimbangkan untuk mengumpulkan lebih banyak data atau melakukan analisis tambahan untuk memperkuat temuan ini.

Chi-Squared Value: 140.24

Nilai Chi-Squared yang tinggi ini menunjukkan ada perbedaan yang signifikan antara frekuensi yang diamati dan frekuensi yang diharapkan dalam tabel kontingensi. Ini mengindikasikan bahwa ada kemungkinan hubungan antara variabel yang diuji.

```python
import pandas as pd
from scipy import stats  # Pastikan untuk mengimpor stats di sini

# Membaca data
df = pd.read_csv('Data/Cleaned/cleaned_data.csv')

# Memilih kolom dengan iloc
service_rating = df.iloc[:, 16]  # Kolom penilaian layanan
loyalty = df.iloc[:, 19]  # Kolom kesetiaan

# Membuat tabel kontingensi
contingency_table = pd.crosstab(service_rating, loyalty)

# Melakukan uji Chi-Squared
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Menampilkan hasil
print("Chi-Squared Value:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
```

Berdasarkan hasil analisis ini, Anda dapat menyimpulkan bahwa tidak ada hubungan yang signifikan antara penilaian layanan di Starbucks dan kesetiaan pelanggan.
