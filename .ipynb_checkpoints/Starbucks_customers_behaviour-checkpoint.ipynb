# Pengenalan

**Deskripsi Dataset:**

Dataset ini terdiri daripada soalan kaji selidik yang dijawab oleh lebih 100 responden mengenai tingkah laku pembelian mereka di Starbucks. Pendapatan dinyatakan dalam Ringgit Malaysia (RM).

**Kandungan**

Dataset mengandungi maklumat demografi pelanggan seperti:

1. Jantina
2. Julat umur
3. Status pekerjaan
4. Julat pendapatan
   
Data ini juga mengandungi maklumat mengenai tingkah laku pembelian pelanggan, termasuk:

Ciri dan kemudahan Starbucks yang menyumbang kepada pilihan mereka.

# Objektif

1. Menentukan Profil Pelanggan Setia: **(Analisis deskriptif)**

- Kenal pasti ciri-ciri demografi pelanggan (seperti umur, pendapatan, dan status pekerjaan) yang paling sering membeli di Starbucks.
- Analisis untuk mengetahui kumpulan demografi mana yang paling cenderung menjadi pelanggan setia.


Umur - Siapakah kumpulan umur yang paling banyak membeli di Starbucks?\
Pendapatan - Apakah kumpulan pendapatan yang sering menjadi pelanggan?\
Status Pekerjaan - Adakah pelanggan yang bekerja sepenuh masa, pelajar, atau freelancer lebih cenderung menjadi pelanggan setia?

2. Mengenal Pasti Faktor Utama Penarik Pelanggan: **(analisis korelasi)**
   
- Analisis ciri-ciri dan kemudahan di Starbucks yang paling menarik perhatian pelanggan, seperti lokasi, suasana, kemudahan Wi-Fi, menu, dan lain-lain.
- Mengukur sejauh mana faktor-faktor ini mempengaruhi keputusan untuk membeli atau berkunjung semula.
  
3. Meramal Tingkah Laku Pembelian Masa Hadapan: **(Forecasting)**

- Bangunkan model ramalan berdasarkan data demografi dan tingkah laku semasa untuk mengetahui kemungkinan seseorang pelanggan akan terus membeli di Starbucks.
- Fokus kepada pelanggan dengan peluang tinggi untuk menjadi pelanggan berulang.

# 2) Analisis Data yang Hilang

```python
#using iloc
# Step 1: Identify rows with missing values in the specific columns using iloc
missing_mask = df.iloc[:, [6, 19]].isnull().any(axis=1)

# Step 2: Use iloc to view the rows with missing values
missing_rows = df[missing_mask]
missing_rows

```

### Menggunakan mode untuk gantikan nilai missing data NAN bagi kategori, supaya analisis yang lebih tepat

```python
# Mengira mod untuk kolum yang mempunyai nilai NaN
mode_6 = df['6. How do you usually enjoy Starbucks?'].mode()[0]  # Mod untuk kolum 6
mode_19 = df['19. How do you come to hear of promotions at Starbucks? Check all that apply.'].mode()[0]  # Mod untuk kolum 19

# Gantikan NaN dengan mod dan simpan dalam DataFrame baru
cleaned_df = df.copy()  # Buat salinan DataFrame asal
cleaned_df['6. How do you usually enjoy Starbucks?'].fillna(mode_6, inplace=True)
cleaned_df['19. How do you come to hear of promotions at Starbucks? Check all that apply.'].fillna(mode_19, inplace=True)

# Semak DataFrame selepas penggantian
cleaned_df
```

```python
cleaned_df.isna().sum()
```

## Objektif 1 = Kenal Pasti Pelanggan Setia

**Identify Loyal Customers**

Based on the questions in your dataset, define the criteria for a "loyal customer." For example:

Question 5: “How often do you visit Starbucks?” – Customers who answered “Daily” or “Weekly” may be considered loyal.
                                                           
Question 20: “Will you continue buying at Starbucks?” – If they answered “Yes,” it shows a willingness to keep purchasing.

```python
# Display unique values for the specific columns by their index positions
unique_values_5 = df.iloc[:, 5].unique()  # Column at index 5
unique_values_20 = df.iloc[:, 20].unique()  # Column at index 20

print("Unique values in column at index 5 (Frequency of visit):", unique_values_5)
print("Unique values in column at index 20 (Will continue buying):", unique_values_20)
```

```python
# Create a new variable 'cleaned_df_with_loyal' that includes the 'Loyal Customer' column
cleaned_df_with_loyal = cleaned_df.copy()  # Make a copy of the cleaned DataFrame

# Define conditions for loyal customers
conditions_loyal = (cleaned_df_with_loyal.iloc[:, 5].isin(["Weekly", "Daily"])) & (cleaned_df_with_loyal.iloc[:, 20] == "Yes")

# Add the 'Loyal Customer' column based on the conditions
cleaned_df_with_loyal['Loyal Customer'] = conditions_loyal.astype(int)  # 1 for loyal, 0 for not loyal

# Display the first few rows to verify
cleaned_df_with_loyal.head()
```

# Visualize data 

```python
sns.catplot(data=cleaned_df_with_loyal, 
             x='2. Your Age', 
             hue='Loyal Customer', 
             kind='count', 
             palette={0: 'red', 1: 'green'})
```

```python
sns.catplot(data=cleaned_df_with_loyal,
             x='2. Your Age',
             y='14. How important are sales and promotions in your purchase decision?',
             hue='Loyal Customer',
             kind='violin',
             palette={0: 'red', 1: 'green'})

# Menampilkan plot
import matplotlib.pyplot as plt
plt.title('Pengaruh Kelayakan Pelanggan terhadap Pentingnya Promosi')
plt.show()
```

```python
sns.catplot(data=cleaned_df_with_loyal,
             x='2. Your Age',
             y='15. How would you rate the ambiance at Starbucks? (lighting, music, etc...)',
             hue='Loyal Customer',
             kind='box',
             palette={0: 'red', 1: 'green'})
```

```python
# Menggunakan bar plot untuk melihat pengedaran kualiti
sns.barplot(data=cleaned_df_with_loyal,
             x='1. Your Gender',
             y='12. How would you rate the quality of Starbucks compared to other brands (Coffee Bean, Old Town White Coffee..) to be:' )

# Menampilkan plot
plt.title('Kualiti Starbucks Berdasarkan Jantina')
plt.show()
```

```python
# Calculate the total number of loyal customers
total_loyal_customers = cleaned_df_with_loyal['Loyal Customer'].sum()
total_loyal_customers
```

```python
# Filter the DataFrame to show only loyal customers
loyal_customers_df = cleaned_df_with_loyal[cleaned_df_with_loyal['Loyal Customer'] == 1]

loyal_customers_df
```

```python
# Save the data
loyal_customers_df.to_csv('loyal_customers_df.csv', index=False)
```

```python
sns.countplot(data=loyal_customers_df, x='1. Your Gender', hue='Loyal Customer')
plt.title('Kedudukan Pelanggan Setia Berdasarkan Jantina')
plt.xlabel('Jantina')
plt.ylabel('Jumlah Pelanggan')
plt.legend(title='Pelanggan Setia', labels=['Tidak Setia (0)', 'Setia (1)'])
plt.show()
```
